```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
def plot_data(df, x_label, y_label, bin_size=1):
    """
    This function takes a DataFrame and plots the data.
    :param df: DataFrame containing the data to plot
    :param x_label: Label for the x-axis
    :param y_label: Label for the y-axis
    """
    resulty = np.array(df[y_label])
    resulty = resulty[:(resulty.size // bin_size * bin_size)].reshape(-1, bin_size).mean(axis=1)

    resultx = np.array(df[x_label])
    resultx = resultx[:(resultx.size // bin_size * bin_size)].reshape(-1, bin_size).mean(axis=1)

    plt.figure(figsize=(10, 6))
    plt.plot(resultx, resulty, linestyle='-')
    plt.title(f'{y_label} vs {x_label}')
    plt.xlabel(x_label)
    plt.ylabel(y_label)
    plt.grid()
    plt.show()
```

```python
df = pd.read_csv('logs/1000000_cycle.csv')
plot_data(df, 'timestep', 'reward', bin_size=10000)
```
